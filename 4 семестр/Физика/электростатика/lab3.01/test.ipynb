```python
import numpy as np
import pandas as pd

# Исходные данные
y = np.array([2, 4, 6, 8, 10, 12, 14, 16, 18])
x = np.array([2, 4.1, 6.2, 9.1, 12.2, 16.8, 19.3, 22.1, 24, 26.2, 28])
phi = np.array([1.76, 2.76, 3.76, 4.76, 5.76, 6.76, 7.76, 8.76, 9.76, 10.76, 11.64])

# Расчет разности потенциалов (phi_i - phi_j)
# np.abs гарантирует, что значение напряженности будет положительным
delta_phi = np.abs(np.diff(phi))

# Массив расстояний l_ij (в см) между точками измерения на одной линии.
# Замените массив l_ij на ваши реальные расстояния. 
# В качестве примера заполняем его единицами (1 см).
l_ij = np.ones(len(delta_phi)) 

# Расчет средней напряженности <E_ij> = delta_phi / l_ij
E = delta_phi / l_ij
# Генерация заголовков столбцов из массива phi
columns = [f"{phi[i]}-{phi[i+1]}" for i in range(len(phi)-1)]

# Формирование итоговой таблицы (DataFrame)
# Значения напряженности дублируются для каждой координаты Y
df = pd.DataFrame([E] * len(y), index=y, columns=columns)
df.index.name = "Y, см"

print("Таблица напряженностей E, В/см:\n")
print(df.round(3)) # Округление до 3 знаков после запятой
```
