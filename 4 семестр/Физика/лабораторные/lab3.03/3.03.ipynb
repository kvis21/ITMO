```python
import plotly.graph_objects as go
import numpy as np

# 1. Данные
Ic = np.array([0.02, 0.04, 0.06, 0.08, 0.10, 0.13, 0.16, 0.19, 0.22, 0.25, 
               0.28, 0.31, 0.34, 0.37, 0.40, 0.43, 0.46, 0.48, 0.50])

data_dict = {
    '9.0 В': np.array([0.205, 0.2051, 0.2051, 0.2051, 0.2051, 0.2055, 0.2051, 0.2028, 
                      0.1911, 0.1331, 0.0951, 0.0827, 0.067, 0.0544, 0.0458, 0.0395, 
                      0.0355, 0.0331, 0.0306]),
    '12.0 В': np.array([0.2502, 0.2503, 0.2502, 0.2502, 0.2496, 0.2503, 0.2509, 0.2497, 
                       0.2385, 0.1776, 0.1303, 0.1036, 0.0879, 0.0763, 0.0635, 0.0566, 
                       0.0475, 0.0444, 0.0414]),
    '13.5 В': np.array([0.3392, 0.3392, 0.3378, 0.3380, 0.3382, 0.3405, 0.3411, 0.3409, 
                        0.3337, 0.2523, 0.1976, 0.1530, 0.1379, 0.1147, 0.0978, 0.0876, 
                        0.0777, 0.0712, 0.0652])
}

fig = go.Figure()
colors = {'9.0 В': 'blue', '12.0 В': 'green', '13.5 В': 'red'}

for label, Ia in data_dict.items():
    # Основной график
    fig.add_trace(go.Scatter(x=Ic, y=Ia, name=label, line=dict(color=colors[label])))

    # Поиск точки максимального спада (минимум производной)
    dIa = np.diff(Ia) / np.diff(Ic)
    idx = np.argmin(dIa) 
    
    # Параметры для касательной в точке максимального наклона
    x0, y0 = Ic[idx], Ia[idx]
    slope = dIa[idx]
    
    # Уровень плато (максимальный ток в начале)
    y_plateau = np.mean(Ia[:5])
    
    # Расчет I_кр (пересечение y_plateau = slope*(x - x0) + y0)
    I_kr = (y_plateau - y0) / slope + x0
    
    # Отрисовка касательной
    x_range = np.array([I_kr - 0.05, x0 + 0.05])
    y_tangent = slope * (x_range - x0) + y0
    fig.add_trace(go.Scatter(x=x_range, y=y_tangent, mode='lines', 
                             line=dict(color=colors[label], dash='dot'), 
                             showlegend=False, opacity=0.5))
    
    # Отметка I_кр вертикальной линией
    fig.add_vline(x=I_kr, line_width=1, line_dash="dash", line_color=colors[label])
    
    # Точка пересечения на графике
    fig.add_trace(go.Scatter(x=[I_kr], y=[y_plateau], mode='markers',
                             marker=dict(size=10, symbol='x-thin', line=dict(width=2, color='black')),
                             name=f'I_кр ({label}) = {I_kr:.3f} мА'))

fig.update_layout(
    title="Определение критического тока I_кр методом касательных",
    xaxis_title="Ток соленоида Ic, мА",
    yaxis_title="Анодный ток Ia, мА",
    hovermode="closest",
    template="plotly_white"
)

fig.show()
```
