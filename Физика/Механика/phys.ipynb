```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

# Данные сгруппированные по рискам (1-6 положений утяжелителей)
# Каждая строка - данные для одной риски
epsilon_data = np.array([
    [2.89, 4.97, 6.95, 8.29],  # 1 риска
    [2.05, 3.86, 4.80, 6.42],  # 2 риска
    [1.62, 2.93, 3.79, 5.23],  # 3 риска
    [1.10, 2.03, 3.20, 4.13],  # 4 риска
    [0.88, 1.62, 2.77, 2.78],  # 5 риска
    [0.75, 1.38, 1.89, 2.35]   # 6 риска
])

M_data = np.array([
    [0.06, 0.11, 0.16, 0.20],
    [0.06, 0.11, 0.16, 0.21],
    [0.06, 0.11, 0.16, 0.21],
    [0.06, 0.11, 0.16, 0.21],
    [0.06, 0.11, 0.16, 0.21],
    [0.06, 0.11, 0.16, 0.21]
])

# Цвета как во втором графике
colors = ['blue', 'green', 'red', 'orange', 'purple', 'brown']
risk_labels = [f'{i+1} риска' for i in range(6)]

plt.figure(figsize=(8, 6))

# Построение графиков с линейной аппроксимацией для каждой риски
for i in range(6):
    epsilon = epsilon_data[i]
    M = M_data[i]
    
    # Точки экспериментальных данных
    plt.scatter(epsilon, M, color=colors[i], s=120, 
                edgecolors='black', linewidth=1, alpha=0.8,
                label=f'{risk_labels[i]}', zorder=5)
    
    # Линейная регрессия (аппроксимация)
    slope, intercept, r_value, p_value, std_err = stats.linregress(epsilon, M)
    
    # Линия аппроксимации
    x_fit = np.array([min(epsilon) * 0.9, max(epsilon) * 1.1])
    y_fit = slope * x_fit + intercept
    
    plt.plot(x_fit, y_fit, color=colors[i], linewidth=2.5, 
             linestyle='-', alpha=0.7, zorder=4)

# Настройки графика
plt.xlabel('Угловое ускорение ε, с⁻²', fontsize=14, fontweight='bold')
plt.ylabel('Момент силы натяжения M, Н·м', fontsize=14, fontweight='bold')
plt.title('Зависимость момента силы натяжения нити М от углового ускорения крестовины ε', 
          fontsize=12, fontweight='bold')

plt.grid(True, linestyle='--', alpha=0.6, which='both')
plt.legend(fontsize=11, loc='upper left', frameon=True, framealpha=0.95)

# Установка пределов осей
plt.xlim(0, max(epsilon_data.flatten()) * 1.15)
plt.ylim(0, max(M_data.flatten()) * 1.15)

# Добавление сетки
ax = plt.gca()
ax.minorticks_on()
ax.grid(True, which='major', linestyle='-', linewidth=0.5, alpha=0.7)
ax.grid(True, which='minor', linestyle=':', linewidth=0.5, alpha=0.4)

plt.tight_layout()
plt.show()

# Дополнительная информация
print("=" * 65)
print("ЛИНЕЙНАЯ АППРОКСИМАЦИЯ M(ε) ДЛЯ РАЗНЫХ ПОЛОЖЕНИЙ УТЯЖЕЛИТЕЛЕЙ")
print("=" * 65)
print("Формула аппроксимации: M = I·ε + Mтр")
print("где I - момент инерции, Mтр - момент сил трения")
print("=" * 65)

for i in range(6):
    slope, intercept, r_value, p_value, std_err = stats.linregress(epsilon_data[i], M_data[i])
    print(f"\n{risk_labels[i].upper()}:")
    print(f"  Уравнение: M = {slope:.5f}·ε + {intercept:.5f}")
    print(f"  Коэффициент детерминации R² = {r_value**2:.4f}")
    print(f"  Момент инерции (I) ≈ {slope:.5f} Н·м·с²")
    print(f"  Момент сил трения (Mтр) ≈ {intercept:.5f} Н·м")
print("=" * 65)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Данные из Таблицы 4
# Квадрат расстояния R^2, м^2
R2 = np.array([0.005929, 0.010404, 0.016129, 0.023104, 0.031329, 0.040804])

# Момент инерции I, кг·м^2 (предположительно)
I = np.array([0.03, 0.03, 0.04, 0.05, 0.06, 0.09])

# Рассчет линейной регрессии
coeffs = np.polyfit(R2, I, 1)
poly = np.poly1d(coeffs)
I_fit = poly(R2)

# Уравнение линии тренда
slope = coeffs[0]
intercept = coeffs[1]
equation = f'I = {slope:.3f}·R² + {intercept:.3f}'

plt.figure(figsize=(10, 6))

# Построение экспериментальных точек
plt.scatter(R2, I, color='red', s=100, zorder=5, 
           label='Экспериментальные данные')

# Построение линии тренда
plt.plot(R2, I_fit, color='blue', linewidth=2, 
        label=f'Линейная аппроксимация: {equation}')

# Настройки графика
plt.xlabel('Квадрат расстояния R², м²', fontsize=12)
plt.ylabel('Момент инерции I, кг·м²', fontsize=12)
plt.title('Зависимость момента инерции крестовины от квадрата расстояния\nмежду осью вращения и центром утяжелителя', 
          fontsize=14, pad=15)
plt.grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()

# Сохранение графика
#plt.savefig('graph2_I_vs_R2.png', dpi=300)
plt.show()

# Дополнительная информация
print("Анализ линейной зависимости:")
print(f"Угловой коэффициент (наклон): {slope:.4f} кг")
print(f"Свободный член: {intercept:.4f} кг·м²")
print(f"Коэффициент детерминации R²: {np.corrcoef(R2, I)[0,1]**2:.4f}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Данные из таблицы 3
T = [1.68, 1.76, 1.86, 1.97, 2.12, 2.26]  # период, с
I = [0.032, 0.036, 0.040, 0.046, 0.053, 0.060]  # момент инерции

# Вычисление квадрата периода
T_squared = [t**2 for t in T]

# Линейная регрессия
slope, intercept = np.polyfit(I, T_squared, 1)
I_fit = np.linspace(min(I), max(I), 100)
T_squared_fit = slope * I_fit + intercept

# Построение графика
plt.figure(figsize=(6, 4))
plt.plot(I, T_squared, 'bo', linewidth=2, markersize=8)
plt.plot(I_fit, T_squared_fit, 'r-', linewidth=1)

plt.xlabel('I, кг·м²', fontsize=12)
plt.ylabel('T², с²', fontsize=12)
plt.title('Зависимость квадрата периода от момента инерции T²(I)', fontsize=10)

plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
